```python
import pandas as pd

def detect_outliers(df, columns_to_check, threshold=1.5):
   
    outlier_report = pd.DataFrame(columns=['Variable', 'Outliers', 'Observation', 'Reasoning'])
    
    for column in columns_to_check:
        if column in df.columns:
            data = df[column]
            
            # Calculate the IQR
            Q1 = data.quantile(0.25)
            Q3 = data.quantile(0.75)
            IQR = Q3 - Q1
            
            # Define the upper and lower bounds for outliers
            lower_bound = Q1 - threshold * IQR
            upper_bound = Q3 + threshold * IQR
            
            # Detect outliers
            outliers = data[(data < lower_bound) | (data > upper_bound)]
            
            # Create an observation and reasoning
            observation = "No outliers detected."
            reasoning = "Data points outside of the IQR range (1.5 times IQR) are considered outliers."
            if len(outliers) > 0:
                observation = f"{len(outliers)} outliers detected."
                reasoning = f"{len(outliers)} data points are outside of the IQR range (1.5 times IQR)."
            
            # Append the result to the report DataFrame
            outlier_report = outlier_report.append({
                'Variable': column,
                'Outliers': len(outliers),
                'Observation': observation,
                'Reasoning': reasoning
            }, ignore_index=True)
    
    return outlier_report



outlier_report = detect_outliers(df2, numerical_column_names)
print(outlier_report)
```

```python
import pandas as pd

def find_continuous_variables(dataframe, exclude=[]):
    continuous_variables = []
    for column in dataframe.columns:
        if column not in exclude:
            dtype = dataframe[column].dtype
            
            # Check if the column is numeric and not categorical
            if pd.api.types.is_numeric_dtype(dtype):
                continuous_variables.append(column)
    
    return continuous_variables


continuous_vars = find_continuous_variables(df2)
print("Continuous Variables:", continuous_vars)





```
